This function performs a policy analysis using the experiments above. For each experiments provided, it runs multiple simulation iterations of the `EvacuationDec` model using the `batch_run` function. The results are collected in a DataFrame and saved as a CSV file, allowing for later analysis of how changes in parameters influence outcomes. This approach helps to identify which parameters have the most significant effects on model results.

<span style="color:red"> The results are saved, so the cell can be skipped if rerunning the analysis is not required.  </span>.

```python
def sensitivity_analysis(index,parameter):
    print(f"Run for {parameter}")
    # We want to do a sensitivity analysis for all variables in the parameter_list
    # Run the batchrunner for the variable
    # for value in parameter[key]:
    result_batch = batch_run(
    EvacuationDec,
    parameters=parameter,
    iterations=150,
    max_steps=90,
    number_processes= 12,
    data_collection_period=1,
    display_progress=True,
    )
    df_sensitivity = pd.DataFrame(result_batch)
    df_sensitivity.to_csv(f"archives/policy_runs/{index}.csv")
    del result_batch, df_sensitivity

for index,parameter in enumerate(para_list) :
    results = sensitivity_analysis(index,parameter)
```

```python

for column_index in range(len(csv_files)):

    data = pd.read_csv(f"archives/policy_runs/{column_index}.csv")
    data_grouped = data[["evac_agents","average_evac_time","Step"]].groupby(by="Step").mean()
    total_EvacDec = data_grouped["evac_agents"].sum()
    b_25 = float(total_EvacDec * 0.25)
    b_75 = float(total_EvacDec * 0.75)
    # print(b_25,b_75)
    cum = 0
    lever_25 = False
    lever_75 = False
    for index, i in enumerate(data_grouped["evac_agents"]):
        # print(i,b_25)
        cum += i
        # print(cum)
        if cum > b_25 and lever_25 == False:
            step_25 = index -1
            lever_25 = True
        if cum > b_75 and lever_75 == False:
            step_75 = index
            lever_75 = True
    average_evacdec = data_grouped[data_grouped.index ==data_grouped.index.max()]["average_evac_time"].values[0]
    spread_metric = step_75 -step_25
    data_max = data[data["Step"] ==data["Step"].max()]
    indecesive_agents = data[data["Step"]== data["Step"].max()][["phase_0","phase_1","phase_2"]].sum().sum()/150

    metric_df[metric_df.columns[column_index]] = [average_evacdec,spread_metric,indecesive_agents]
metric_df.index = ["Mean Decision time","Decision Spread", "Indecesive Agents"]
```
